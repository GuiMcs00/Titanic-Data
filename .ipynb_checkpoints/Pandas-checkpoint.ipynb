Filtros:

```python
# podemos realizar buscas com operadores lÃ³gicos
base[base.Fare > 500]
```

```python
base[base.Fare < 5]
```

```python
# podemos realizar buscas com expressÃµes lÃ³gicas
base[(base.Parch > 1) & (base.SibSp > 1)]
```

```python
# loc permite realizar busca da mesma forma acima, mas filtrando colunas de uma forma prÃ¡tica
base.loc[(base.Parch > 1) | (base.SibSp > 1), ['PassengerId', 'Parch', 'SibSp']].head()
```

```python
# iloc Ã© usado com linhas
base.iloc[30:40]
```

```python
# tambÃ©m podemos filtrar colunas especÃ­ficas
base.iloc[30:40, 3:6]
```

```python
# podemos ordenar os valores
base.loc[(base.Age >= 18) & (base.Age <= 25)].sort_values(['Age', 'Name'],ascending=True)
```
